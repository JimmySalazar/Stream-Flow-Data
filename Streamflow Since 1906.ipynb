## Basic Plot With No Fourier Transform

### 1906 - 2005

```python
import numpy as np
import matplotlib.pyplot as plt

# Define Empty Lists For Data

x = [] # Years 1906 - 2005
y = [] # Stream Flow Data 1906 - 2005

# Load Data & Append to Empty Lists

for num in np.loadtxt('Years_Observed.txt'):
    x.append(num)
    
for num in np.loadtxt('Flow_Observed.txt'):
    y.append(num)
    
# Adjust Size and DPI of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Graph Title

plt.title("Tree Ring Data From 1906 to 2005")

# X & Y Axis Labels

plt.xlabel("Year")
plt.ylabel("Streamflow of Colorado River (Millions Acre/Ft)")

# Limit Graph to 1906-2005 X Values

plt.xlim([1906, 2005])

plt.plot(x, y, label = 'Actual Observed Data')
plt.legend()
plt.show()
```

```python
fft = np.fft.fft(y)

# Adjusting size of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Limit Graph to 1906-2005 X Values

plt.xlim([0.01, .5])
plt.ylim([0, 1e8])

z = np.abs(fft)
n = 100
freq = np.fft.fftfreq(n, d=1)

# X & Y Axis Labels

plt.xlabel("Frequency")
#plt.xticks([])

print(freq)
plt.plot(np.abs(freq), z)
plt.show()
```
