## Basic Plot With No Fourier Transform

### 1906 - 2005

```python
import numpy as np
import matplotlib.pyplot as plt

# Define Empty Lists For Data

x = [] # Years 1906 - 2005
y = [] # Stream Flow Data 1906 - 2005

# Load Data & Append to Empty Lists

for num in np.loadtxt('Years_Observed.txt'):
    x.append(num)
    
for num in np.loadtxt('Flow_Observed.txt'):
    y.append(num)
    
# Adjust Size and DPI of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Graph Title

plt.title("Tree Ring Data From 1906 to 2005")

# X & Y Axis Labels

plt.xlabel("Year")
plt.ylabel("Streamflow of Colorado River (Millions Acre/Ft)")

# Limit Graph to 1906-2005 X Values

plt.xlim([1906, 2005])

plt.plot(x, y, label = 'Actual Observed Data')
plt.legend()
plt.show()
```

## Fourier Transform of Streamflow Data

#### This shows the array created by the FFT made up of real and imaginary numbers

```python
np.fft.fft(y)
```

## Graph of Discreet Fourier Transform Imaginary Numbers

```python
# Assigning the Fourier Transform to a variable

fft = np.fft.fft(y)

# Adjusting size of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Limit Graph to 1906-2005 X Values

plt.xlim([1906, 2005])

plt.plot(x, fft.imag)
plt.show()
```

## Graph of Discreet Fourier Transform Real Numbers

```python
# Assigning the Fourier Transform to a variable

fft = np.fft.fft(y)

# Adjusting size of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Limit Graph to 1906-2005 X Values

plt.xlim([1906, 2005])

plt.plot(x, fft.real)
plt.show()
```
